```python
import matplotlib.pyplot as plt
```

```python
hE = []
sumaE = 0
cantE = 0
with open('m30_2024_datos_histograma_atrib2_dataEstructura.txt') as fp:
    for line in fp:
        data = line.split()
        v = int(data[0])
        x = int(data[1])
        hE.append(x)
        sumaE += (x*v)
        cantE += x
```

```python
plt.plot(hE)
plt.title("Histograma de datos [Estructura]")
plt.savefig('m30_2024_datos_histograma_atrib2_dataEstructura.png', bbox_inches='tight')
plt.grid(axis='y')
#plt.yscale('log')
#plt.legend()
plt.show()
plt.close()
```

```python
hO = []
sumaO = 0
cantO = 0
with open('m30_2024_datos_histograma_atrib2_dataOriginal.txt') as fp:
    for line in fp:
        data = line.split()
        v = int(data[0])
        x = int(data[1])
        hO.append(x)
        sumaO += (x*v)
        cantO += x
```

```python
plt.plot(hO)
plt.title("Histograma de datos [Original]")
plt.savefig('m30_2024_datos_histograma_atrib2_dataOriginal.png', bbox_inches='tight')
plt.grid(axis='y')
#plt.yscale('log')
#plt.legend()
plt.show()
plt.close()
```

```python
print("Estadsticas")
print("param\tprevEs\t\tpostEs")
print("cant\t" + str(cantO) + "\t\t" + str(cantE))
print("suma\t" + str(sumaO) + "\t" + str(sumaE))

promE = round(sumaE / cantE, 2)
promO = round(sumaO / cantO, 2)
print("prom\t" + str(promO) + "\t\t" + str(promE))
```
