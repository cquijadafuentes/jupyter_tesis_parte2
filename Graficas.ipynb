```python
# Importanciones

import matplotlib.pyplot as plt
import numpy as np
```

```python
species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
plt.close()
```

```python
def graficaSTAI(labels, data, titulo, archivo):
    x = np.arange(len(labels))  # the label locations
    barwidth = 0.2
    m = 0
    fig, ax = plt.subplots(layout='constrained')
    for attribute, measurement in data.items():
        offset = barwidth * m
        rects = ax.bar(x + offset, measurement, barwidth, label=attribute)
        ax.bar_label(rects, padding=3)
        m += 1
    ax.set_ylabel('STAI')
    ax.set_title(titulo)
    ax.set_xticks(x + barwidth, labels )
    ax.legend(loc='upper left', ncols=3)
    ax.set_ylim(-1, 1)
    plt.savefig(archivo, bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
labels = ("Intensidad", "Ocupación", "Carga", "Velocidad Media")
data = {
	'PM1xxx1': (0.36, 0.40, 0.45, 0.40),
	'PM1xxx2': (0.56, 0.37, 0.18, 0.29),
	'PM2xxx1': (0.19, 0.43, 0.52, 0.18),
	'PM2xxx2': (0.41, 0.17, 0.00, -0.12)
}
titulo = 'Spatio-Temporal Autocorrelation Index (STAI) Sensores Madrid'
archivo = "graficas/STAI_sensores_madrid_I-E_T-L"
graficaSTAI(labels, data, titulo, archivo)
```
