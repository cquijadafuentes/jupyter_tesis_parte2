```python
import folium
import pandas as pd

# Leer los datos desde un archivo CSV
data = pd.read_csv('pmed_ubicacion_11-2024_M30.csv', sep=';', quotechar='"')

print(data.head())
print("Total sensores:", data.shape[0])
```

```python
# Crear un mapa centrado en Madrid
m1 = folium.Map(location=[40.4168, -3.7038], zoom_start=12)

# Agregar los marcadores
for index, row in data.iterrows():
    if(row['cod_cent'][2]=='1'): 
        folium.Marker([row['latitud'], row['longitud']], popup=row['cod_cent'], icon=folium.Icon(color='green')).add_to(m1) # Calzada Interior
    elif(row['cod_cent'][2]=='2'):
        folium.Marker([row['latitud'], row['longitud']], popup=row['cod_cent'], icon=folium.Icon(color='red')).add_to(m1) # Calzada Exterior
    elif(row['cod_cent'][2]=='3'):
        folium.Marker([row['latitud'], row['longitud']], popup=row['cod_cent'], icon=folium.Icon(color='orange')).add_to(m1) # Carretera de Salida
    elif(row['cod_cent'][2]=='4'):
        folium.Marker([row['latitud'], row['longitud']], popup=row['cod_cent'], icon=folium.Icon(color='blue')).add_to(m1) # Carretera de Entrada

# Mostrar el mapa
m1
```

Mapa de las carreteras de entradas/salidas de Madrid

```python
# Crear un mapa centrado en Madrid
m2 = folium.Map(location=[40.4168, -3.7038], zoom_start=12)

# Agregar los marcadores
pCal = 2
for index, row in data.iterrows():
    if(row['cod_cent'][pCal]=='3' or row['cod_cent'][pCal]=='4'):
        folium.Marker([row['latitud'], row['longitud']], popup=row['cod_cent']).add_to(m2)

# Mostrar el mapa
m2
```

Mapa de las que no son Tronco ni Servicio

```python
# Crear un mapa centrado en Madrid
m3 = folium.Map(location=[40.4168, -3.7038], zoom_start=12)

# Agregar los marcadores
for index, row in data.iterrows():
    if(row['cod_cent'][2]!='1' and (row['cod_cent'][6]!='1' or row['cod_cent'][6]!='2')):
        # Interior y (Tronco o Laterial) == GREEN
        folium.Marker([row['latitud'], row['longitud']], popup=row['cod_cent'], icon=folium.Icon(color='green')).add_to(m3)
    elif(row['cod_cent'][2]!='2' and (row['cod_cent'][6]!='1' or row['cod_cent'][6]!='2')):
        # Exterior y (Tronco o Laterial) == DARKRED
        folium.Marker([row['latitud'], row['longitud']], popup=row['cod_cent'], icon=folium.Icon(color='darkred')).add_to(m3)

# Mostrar el mapa
m3
```
